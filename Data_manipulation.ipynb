# Data joining and merging

```python
df1=pd.merge(reactions,reactType,how='inner')
```

```python
df1.head()
```

```python
df2=pd.merge(session,user,how='inner')
```

```python
df2.head()
```

```python
df3=pd.merge(df2,loc,how='inner')
```

```python
df3.head()
```

```python
df4=pd.merge(df3,profile,how='inner')
```

```python
df4.head()
```

```python
df5=pd.merge(df4,content,how='inner')
```

```python
df5.head()
```

```python
df=pd.merge(df1,df5,on='User ID')
```

```python
df.head()
```

```python
#renaming type_x which belongs to Reaction data as ReactionType
df.rename(columns={'Type_x':'ReactionType'},inplace=True)
```

```python
df.head()
```

```python
#renaming content type_y to contentType
df.rename(columns={'Type_y':'ContentType'},inplace=True)
```

```python
df.shape
```

# Data Cleaning

```python
for c in df.columns:
    miss=df[c].isnull().sum()
    if miss > 0:
        print("{} has {} missing Value(s)".format(c,miss))
    else:
        print('{} has NO missing value(s)!'.format(c))
```

```python
#URL has a lot of missing data (8797) and we wont need the url column and other columns from the problem statement given
data=df.drop(columns=['URL','Content ID_x','Content ID_y','User ID','Email','Name','Interests'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
#changing data types
data['ReactionType'] = data['ReactionType'].astype('str')
data['Sentiment'] = data['Sentiment'].astype('str')
data['Device'] = data['Device'].astype('str')
data['Address'] = data['Address'].astype('str')
data['ContentType'] = data['ContentType'].astype('str')
data['Category'] = data['Category'].astype('str')
data['Duration'] = data['Duration'].astype('float')
```

```python
data['Datetime']=pd.to_datetime(df.Datetime)
```
